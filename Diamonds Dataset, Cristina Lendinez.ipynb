## X, Y, Z

Lo que podemos ver es que tenemos 0 valores en las columnas x, y y z. Esto significaría que hay diamantes que no tienen dimensiones. Es una señal de datos defectuosos, por lo que voy a eliminar esos valores.


```python
#Dropping dimensionless features

diamonds = diamonds.drop(diamonds[diamonds['x'] == 0].index)
diamonds = diamonds.drop(diamonds[diamonds['y'] == 0].index)
diamonds = diamonds.drop(diamonds[diamonds['z'] == 0].index)
```

**2) Realizar el Feature Engineering de los datos para posteriormente realizar el modelo (transformar las columnas necesarias) (2 ptos)**

convirtiendo las variables categopricas a numericas

```python
diamonds_num = diamonds.copy()
#Convirtiendo variables categoricas a numericas

cats = ['color', 'cut', 'clarity']

for i in cats:
    diamonds_num[i+'_numeric'] = pd.factorize(diamonds_num[i])[0]
    diamonds_num.drop(i, axis = 1, inplace = True)
```
